```python
for column in numerical_features:

    plt.figure(figsize=(8,2))

    sns.boxplot(
        x=df[column]
    )

    plt.title(f"Boxplot of {column}")

    plt.tight_layout()
    plt.show()
```

```python
for column in numerical_features:

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[column] < lower) | (df[column] > upper)).sum()

    print("="*60)
    print(column)
    print("Lower Bound :", round(lower,2))
    print("Upper Bound :", round(upper,2))
    print("Number of Outliers :", outliers)
```



## Age
- The boxplot shows **no outliers** in the age feature.
- Using the IQR method, the acceptable range is **-9 to 87 years**.
- Since all ages lie between **18 and 64 years**, every observation falls within the acceptable range.
- This indicates that the age data is clean and does not require any treatment.

## BMI
- BMI contains **9 outliers** according to the IQR method.
- These observations correspond to individuals with unusually high BMI values (greater than approximately **47.29**).
- Such values are medically possible and may represent individuals with severe obesity rather than data entry errors.
- Therefore, these observations are considered valid and should be handled carefully instead of being removed immediately.

## Children
- No outliers were detected in the number of children.
- The values naturally range from **0 to 5**, which is completely reasonable.
- Since this is a discrete numerical feature, all observations are retained.

## Insurance Charges
- The target variable (**charges**) contains **139 outliers** according to the IQR rule.
- These observations represent individuals with exceptionally high medical expenses.
- Since insurance claims naturally vary significantly due to serious illnesses, surgeries, chronic diseases, or smoking-related conditions, these large values are expected in real-world healthcare data.
- These observations are not data errors but represent an important part of the population.

# Why are we NOT removing these outliers?

Although the IQR method identifies BMI and insurance charges as outliers, they are **genuine observations rather than incorrect values**.

Removing these records would have several disadvantages:

- It would reduce the size of the dataset.
- The model would lose information about people with very high medical expenses.
- Predictions for expensive insurance cases would become less accurate.
- The dataset would no longer represent real-world insurance data.

For insurance cost prediction, high charges are expected because healthcare expenses can vary dramatically among individuals.

Therefore, these outliers will **not be removed**.

---

# How will we handle them?

Instead of deleting these observations, we will:

- Keep all records in the dataset.
- Apply appropriate feature scaling during preprocessing.
- Use machine learning algorithms that are capable of handling naturally occurring extreme values.
- Evaluate model performance before deciding whether additional transformations (such as log transformation of the target variable) are necessary.

---

# Conclusion

The dataset contains only a few BMI outliers and a significant number of high insurance charge observations. Since these values are realistic and carry important information, they will be retained. Keeping these observations allows the model to learn the full range of insurance costs and produce predictions that better reflect real-world scenarios.
